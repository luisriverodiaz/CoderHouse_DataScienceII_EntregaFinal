# Análisis con Machine Learning

# Preparación del modelo

```python
from sklearn.preprocessing import LabelEncoder
df_train = df

# Convierto las variables Genero, Tipo Cliente, Tipo Viaje y Clase en numéricas, utilizando LabelEncoder
# Creo el codificador
le = LabelEncoder()

# Aplico el codificador a las columnas
df_train['Genero'] = le.fit_transform(df['Genero'])
df_train['Tipo Cliente'] = le.fit_transform(df['Tipo Cliente'])
df_train['Tipo Viaje'] = le.fit_transform(df['Tipo Viaje'])
df_train['Clase'] = le.fit_transform(df['Clase'])
df_train['Satisfaccion'] = le.fit_transform(df['Satisfaccion'])

print(df_train.head())
```

```python
# Verifico los tipos de datos que sean numéricos
print(df_train.dtypes)
```

```python
# A estos efectos vuelvo a convertir los valores a numéricos
# Voy a corregir los tipos de datos, cambiando a a categórico todos los campos que lo ameritan
indices_corregir = list(range(6,20))
df_train[df_train.columns[indices_corregir]] = df_train[df_train.columns[indices_corregir]].astype('int64')
df_train.info()
```

# Entrenamiento del modelo

```python
from sklearn.model_selection import train_test_split

# Defino X (características) e y (variable objetivo)
X = df_train.drop('Satisfaccion', axis=1)  # Elimino la columna 'Satisfaccion' de las características porque será mi variable objetivo
y = df_train['Satisfaccion']

# Divido el dataset en entrenamiento (70%) y prueba (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Veo los primeros 5 ejemplos del conjunto de entrenamiento
print(X_train.head())
```

```python
# Verifico los tipos de datos del data frame de entrenamiento que sean numéricos
print(X_train.dtypes)
```

```python
# Veo los tamaños de los conjuntos
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

```python
# Defino una función para graficar la matriz de confusión
def grafico_matriz_confusion(y_test, y_pred):
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title('Matriz de Confusión')
    plt.ylabel('Real')
    plt.xlabel('Predicho')
    plt.show()
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import cross_val_score

# Creo el modelo Random Forest con limitación en la profundidad de los árboles para evitar sobreajuste
rf_model = RandomForestClassifier(n_estimators=10, max_depth=8, random_state=42, min_samples_split=2)
rf_model.fit(X_train, y_train)

y_pred_train = rf_model.predict(X_train)
y_pred_test = rf_model.predict(X_test)

print("Reporte de clasificación en el conjunto de entrenamiento:")
print(classification_report(y_train, y_pred_train))

print("Reporte de clasificación en el conjunto de prueba:")
print(classification_report(y_test, y_pred_test))


print("--------------------------------------------------------------------------------------------------")

grafico_matriz_confusion(y_test, y_pred_test)

print("--------------------------------------------------------------------------------------------------")

# Obtengo la importancia de las características
importances = rf_model.feature_importances_
features = X.columns

# Muestro las características más importantes
feature_importance = pd.DataFrame({'Feature': features, 'Importance': importances})
print(feature_importance.sort_values(by='Importance', ascending=False))

print("--------------------------------------------------------------------------------------------------")

# Aplico una validación cruzada para evaluar el rendimiento general del modelo
scores = cross_val_score(rf_model, X_train, y_train, cv=5)
print("Cross-validation scores:", scores)
print("Mean cross-validation score:", scores.mean())
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier(max_depth=7, random_state=42, splitter='best', criterion='gini', min_samples_leaf=1, min_samples_split=2)

dt_model.fit(X_train, y_train)

y_pred_train = dt_model.predict(X_train)
y_pred_test = dt_model.predict(X_test)

print("Reporte de clasificación en el conjunto de entrenamiento:")
print(classification_report(y_train, y_pred_train))

print("Reporte de clasificación en el conjunto de prueba:")
print(classification_report(y_test, y_pred_test))


print("--------------------------------------------------------------------------------------------------")

grafico_matriz_confusion(y_test, y_pred_test)

print("--------------------------------------------------------------------------------------------------")

# Obtengo la importancia de las características
importances = dt_model.feature_importances_
features = X.columns

# Muestro las características más importantes
feature_importance = pd.DataFrame({'Feature': features, 'Importance': importances})
print(feature_importance.sort_values(by='Importance', ascending=False))

print("--------------------------------------------------------------------------------------------------")

# Aplico una validación cruzada para evaluar el rendimiento general del modelo
scores = cross_val_score(dt_model, X_train, y_train, cv=5)
print("Cross-validation scores:", scores)
print("Mean cross-validation score:", scores.mean())
```

## XGBoost

```python
import xgboost

xg_model = xgboost.XGBClassifier(objective='binary:logistic', n_estimators=300, seed=42, max_depth=7, random_state=42, use_label_encoder=False, min_child_weight=1, learning_rate=0.01)

xg_model.fit(X_train, y_train)

y_pred_train = xg_model.predict(X_train)
y_pred_test = xg_model.predict(X_test)

print("Reporte de clasificación en el conjunto de entrenamiento:")
print(classification_report(y_train, y_pred_train))

print("Reporte de clasificación en el conjunto de prueba:")
print(classification_report(y_test, y_pred_test))


print("--------------------------------------------------------------------------------------------------")

grafico_matriz_confusion(y_test, y_pred_test)

print("--------------------------------------------------------------------------------------------------")

# Obtengo la importancia de las características
importances = xg_model.feature_importances_
features = X.columns

# Muestro las características más importantes
feature_importance = pd.DataFrame({'Feature': features, 'Importance': importances})
print(feature_importance.sort_values(by='Importance', ascending=False))

print("--------------------------------------------------------------------------------------------------")

# Aplico una validación cruzada para evaluar el rendimiento general del modelo
scores = cross_val_score(xg_model, X_train, y_train, cv=5)
print("Cross-validation scores:", scores)
print("Mean cross-validation score:", scores.mean())
```

# Optimización de Modelos

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

# Optimizo RandomForest
rf = RandomForestClassifier()
param_grid={
            "n_estimators":[10,20,50,100],
            "max_depth":[5,6,8],
            'min_samples_split': [2, 5, 10]
          }

# Realizo el ajuste utilizando GridSearchCV
gscv = GridSearchCV(rf, param_grid, scoring='accuracy', cv=5, verbose=1)
gscv.fit(X_train, y_train)

# Muestro los mejores hiperparámetros
print("Best parameters: ", gscv.best_params_)

# Predecir en el conjunto de prueba con el mejor modelo
y_pred = gscv.best_estimator_.predict(X_test)

# Evaluar el desempeño
accuracy = accuracy_score(y_test, y_pred)
print("Precisión en el conjunto de prueba: ", accuracy)
```

```python
# Optimizo DecisionTreeClassifier
dtc = DecisionTreeClassifier()

param_grid={
            "max_depth":[3,5,7],
            "min_samples_split":[2,5,10],
            "min_samples_leaf":[1,2,4],
            "criterion":["gini", "entropy"],
            "splitter":["best"],
            "random_state":[42]
          }

# Realizo el ajuste utilizando GridSearchCV
gscv = GridSearchCV(dtc, param_grid, scoring='accuracy', cv=5, verbose=1)
gscv.fit(X_train, y_train)

# Muestro los mejores hiperparámetros
print("Best parameters: ", gscv.best_params_)

# Predecir en el conjunto de prueba con el mejor modelo
y_pred = gscv.best_estimator_.predict(X_test)

# Evaluar el desempeño
accuracy = accuracy_score(y_test, y_pred)
print("Precisión en el conjunto de prueba: ", accuracy)
```

```python
# Optimizo XGBClassifier
xgb = xgboost.XGBClassifier()

param_grid = {
    'n_estimators': [100, 200, 300],         # Número de árboles
    'learning_rate': [0.01],                 # Tasa de aprendizaje (eta)
    'max_depth': [3, 5, 7],                  # Profundidad máxima del árbol
    'min_child_weight': [1, 3, 5]            # Controla la complejidad del árbol
}

# Realizo el ajuste utilizando GridSearchCV
gscv = GridSearchCV(xgb, param_grid, scoring='accuracy', cv=5, verbose=1)
gscv.fit(X_train, y_train)

# Muestro los mejores hiperparámetros
print("Best parameters: ", gscv.best_params_)

# Predecir en el conjunto de prueba con el mejor modelo
y_pred = gscv.best_estimator_.predict(X_test)

# Evaluar el desempeño
accuracy = accuracy_score(y_test, y_pred)
print("Precisión en el conjunto de prueba: ", accuracy)
```
